```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
# Load cleaned data
df = pd.read_csv('StudentsPerformance.csv')

df.head(10)
```

```python
df[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']].describe()
```

###  Categorical Summary:
- Female students are slightly more than male students.
- Most students belong to **Race/Ethnicity Group C**.
- The most common parental education level is **"some college"**.
- Most students receive **standard lunch**.
- Majority of students did **not complete** test preparation.


```python
df['gender'].value_counts()
df['race/ethnicity'].value_counts()
df['parental level of education'].value_counts()
```

###  Key EDA Insights:
- Female students scored higher in reading/writing; males in math.
- Students in Group E had highest scores; Group A the lowest.
- Higher parental education linked to better scores.
- Standard lunch students scored higher than free/reduced.
- Test prep completion strongly improves scores.
