# Seaborn

* It is Python library that is used to visualise data.
* It is an extended version of Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- There are many sample datadsets avilable in seaborn for practise
- Using below function we can import those dataseta directly

```python
sns.get_dataset_names()
```

```python
irs=sns.load_dataset('iris')
irs.head()
```

```python
irs.species.unique()
```

```python
irs.species.value_counts()
```

```python
irs.shape
```

# Lineplot

```python
#Lineplot
sns.lineplot(x='sepal_length',y='sepal_width',data=irs,ci=None) # confidence intervale -Is the 
plt.show() #CI -Is range of values within which true popluation parameter or trend is likely to lie
```

```python
sns.lineplot(x='petal_length',y='petal_width',data=irs,ci=None) # confidence intervale -Is the 
plt.show() #CI -Is range of values within which true popluation parameter or trend is likely to lie
```

# Scatter plots

```python
# Scatter plots are used to show the  relationship between two variables.
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=irs,hue='species')
plt.title("Scatter plot")
plt.show()
```

# Bar Chart

```python
# Bar Charts are a used to show how different categories compare to each other.
```

```python
sns.barplot(x='species',y='sepal_length',data=irs,hue='species')
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=irs)
plt.show()
```

```python
sns.barplot(x='species',y='sepal_width',data=irs)
plt.show()
```

```python
sns.countplot(x='species',data=irs,hue='species')
plt.show()
```

# Boxplot

```python
sns.boxplot(x='sepal_width',data=irs)
plt.show()
```

```python
sns.boxplot(x='sepal_length',data=irs)
plt.show()
```

```python
sns.boxplot(x='species',y='sepal_width',data=irs,hue='species')
plt.show()
```

# Histogram

* Histograms are used to show the distribution of data. They're like a bar chart, but the bars are connected together

```python
sns.histplot(x='sepal_width',data=irs,bins=15,stat='frequency',color='red',kde=True) # KdE= Kernel Density estimation
```

```python
# KDE
```

```python
sns.kdeplot(x='petal_width',data=irs,color='red',fill=True)
```

```python
sns.kdeplot(x='sepal_width',data=irs,color='red',fill=True)
```

# Heat map

* It is generally used to plot the correlation

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
irs.corr(numeric_only=True)
```

```python
sns.heatmap(irs.corr(numeric_only=True),annot=True)
```

# Stripplot

```python
sns.stripplot(x='species',y='sepal_width',data=irs,color='red')
plt.show()
```
